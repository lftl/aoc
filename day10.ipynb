```python
import re
import matplotlib.pyplot as plt
import os

f = open('Downloads/input10').readlines()

# parse = re.compile('position=<([\-0-9]*), *([\-0-9]*)> *velocity=< *([\-0-9]*), *([\-0-9]*)>.*')
parse = re.compile('position=< *([\-0-9]*), *([\-0-9]*)> *velocity=< *([\-0-9]*), *([\-0-9]*)>.*')

points  = []

    
```

```python

for line in f:
    pieces = parse.match(line).groups()
    points.append([int(pieces[0]), int(pieces[1]), int(pieces[2]), int(pieces[3]), int(pieces[0]),int(pieces[1])])
```

points

```python
cur = [[p[4], p[5]] for p in points]

secs = 0
```

```python
def advance(position, points,step=1):
    for x in range(len(points)):
        position[x][0] += (step * points[x][2])
        position[x][1] += (step * points[x][3])
    return position
```

```python
step = 1
secs = 10450
cur = advance(cur, points, 10450)
for x in range(100):
    secs += step
    cur = advance(cur, points, step)
    plt.scatter(
        [x[0] for x in cur],
        [y[1] for y in cur]
    )
    plt.gca().invert_yaxis()

    plt.gcf().set_size_inches(10,1)
    plt.show()
    print(secs)
```
